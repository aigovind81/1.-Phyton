```python
# A program processes a list of numbers and needs to find the largest value.Write logic to identify and return the largest number from a given list.
```

```python
def find_largest(numbers):
    if len(numbers)==0:
        return None
    largest=numbers[0]
    for num in numbers:
        if num>largest:
            largest=num
    return largest
numbers=[5, 15, 45, 75, 99]
result=find_largest(numbers)
print("Largest number:", result)
```

```python
#9.	Scenario: A calendar system verifies whether a given year is a leap year based on standard leap year rules. Write logic to determine whether a given year is a leap year.
```

```python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Check multiple years
years = [1800, 1900, 2000, 2100, 2101, 2109, 2200]

print("Leap Year Checker")
print("=" * 45)
print(f"{'Year':<10} | {'Leap Year?':<12} | {'Reason'}")
print("=" * 45)

for year in years:
    result = "✓ Yes" if is_leap_year(year) else "✗ No"
    
    if year % 400 == 0:
        reason = "Divisible by 400"
    elif year % 100 == 0:
        reason = "Divisible by 100 (not 400)"
    elif year % 4 == 0:
        reason = "Divisible by 4 (not 100)"
    else:
        reason = "Not divisible by 4"
    
    print(f"{year:<10} | {result:<12} | {reason}")
```
