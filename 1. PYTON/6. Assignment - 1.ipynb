```python
lists = [10, 20, 26, 29, 30, 33, 36, 60]
```

```python
for weight in lists:
    if(weight < 18.5):
        print("Underweight")
    elif(weight < 25):
        print("Normal range")
    elif(weight < 30):
        print("Overweight")
    elif(weight < 35):
        print("Obese I")
    elif(weight < 40):
        print("Obese II")
    else:
        print("Obese III")
```
