```python
# Print odd Number in the list
```

```python
numbers = (20, 10, 16, 19, 25, 1, 276, 188)
print(numbers)

for num in numbers:
    if num % 2 !=0:
        print(f"{num} is odd")
```

```python
# Print even Number in the list
```

```python
numbers=[20, 10, 16, 19, 25, 1, 276, 188]
print(tuple(numbers))

for num in numbers:
    if num % 2 !=0:
        print(f"{num} is odd")
```

```python
numbers=[20, 10, 16, 19, 25, 1, 276, 188]
print(tuple(numbers))

for num in numbers:
    if num % 2 == 0:
        print(f"{num} is even")
```
