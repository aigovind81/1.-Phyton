```python
def Agecategory():
        if(age<18):
            print("Children")
            cate="Children"
        elif(age<40):
            print("Adult")
            cate="Adult"
        elif(age<68):
            print("Citizen")
            cate="Citizen"
        else:
            print("Senior Citizen")
            cate="Senior Citizen"
        return cate
```

```python
age=int(input("Enter th Age:"))
agecate=Agecategory()
```

```python
print(agecate)
```
